```python
import numpy as np
from chromauto.ch_parser import CHFile
from chromauto.utils import normalize_baseline

from file_types_examples.eo_fid_integration import EO_Integrate
```

```python
blank_long_file = CHFile('./file_types_examples/eo_blank.ch')

#first_points = np.array(blank_long_6225_fid.values)[:5] - np.array(da.values)[:5]
#avg_fp = np.average(first_points)

base_bx = np.array(blank_long_file.times())
base_by = np.array(blank_long_file.values)

example_data_file = CHFile('./file_types_examples/eo_data.ch')

rx = np.array(example_data_file.times())
ry = np.array(example_data_file.values)
print(example_data_file.metadata)
```

```python
eo_settings = {
    'data_y_win_length': 111,
    'blank_y_win_length': 333,
    'savgol_polyorder': 7,
    'peak_min_dist': 20,
    'peak_max_peak': 3,
    'peak_thres_range': (.5, 10, .001),
    'peak_hard_thres': .00003,
    'norm_thres_range': (20,30),
    'norm_min_dist': 3000, 
    'norm_polyorder': 5,
    'minimum_area': 0.03,
    'split_thres_range':(.01, 5, .001),
    'split_max_peak': 2,
    'split_min_dist': 20,
    'split_min_area': 0.005,
    'min_walking_vert': 0.08,
    'split_win_length': 31,
    'split_polyorder': 5,
    'show_file': True,
    'spec_range': [{'range': (11, 13), 'thres': .0003, 'area': .003}, ]
}

blank_x, blank_y = normalize_baseline(base_bx, base_by, rx, ry, **eo_settings)

EO_Integrate(rx, ry,blank_x,blank_y, './file_types_examples/out5.example',**eo_settings)
```
